# Part I : EDA - Exploratory Data Analysis & Pre-processing

The given dataset seems to be a grocery sales dataset containing information about various products sold by a grocery store. To perform an exploratory data analysis (EDA) on this dataset, we can perform the following tasks:

- Data Import: Import the dataset into a statistical software tool such as Python or R.
- Data Cleaning: Check the dataset for any missing or incorrect data and clean the dataset accordingly. This may involve removing or imputing missing data or correcting any obvious errors.
Data Description: Generate summary statistics such as mean, median, and standard deviation for each column of the dataset. This will help in understanding the distribution of data in each column.
- Data Visualization: Create various visualizations such as histograms, box plots, scatter plots, and heatmaps to understand the relationships and trends between the different variables in the dataset. For example, we can create a scatter plot between the "Fresh" and "Milk" variables to see if there is any correlation between them.
- Outlier Detection: Check for any outliers in the dataset and determine whether they are valid or erroneous data points.
- Correlation Analysis: Calculate the correlation between different variables in the dataset to determine which variables are highly correlated and which ones are not. For example, we can calculate the correlation between "Grocery" and "Detergents_Paper" to see if there is any relationship between these two variables.
- Data Transformation: If necessary, transform the data by standardizing or normalizing the variables to make them comparable across different scales.
- Feature Selection: Identify the most important features or variables that contribute the most to the overall variance in the dataset. This can be done using various feature selection techniques such as principal component analysis (PCA) or random forest regression.

```python
#supress warnings
import warnings
warnings.filterwarnings("ignore")

#import libraries
import pandas as pd
data = pd.read_csv("Wholesale_Data.csv")
data.head

  
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

class DataImporter:
    def import_data(self, file_path):
        return pd.read_csv(file_path)


class DataCleaner:
    def clean_data(self, data):
        # Handle missing or incorrect data
        cleaned_data = data.dropna()  
        cleaned_data = cleaned_data.reset_index(drop=True)
        return cleaned_data


class DataDescriber:
    def generate_summary_statistics(self, data):
        return data.describe()


class DataVisualizer:
    def create_histogram(self, data, column):
        sns.histplot(data[column])
        plt.title(f"Histogram of {column}")
        plt.show()

    def create_boxplot(self, data, column):
        sns.boxplot(data[column])
        plt.title(f"Boxplot of {column}")
        plt.show()

    def create_scatterplot(self, data, x_column, y_column):
        sns.scatterplot(data=data, x=x_column, y=y_column)
        plt.title(f"Scatter plot: {x_column} vs {y_column}")
        plt.show()

    def create_heatmap(self, data):
        correlation_matrix = data.corr()
        sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
        plt.title("Correlation Heatmap")
        plt.show()


class OutlierDetector:
    def detect_outliers(self, data):
        # Perform outlier detection
        # use z-score method
        z_scores = np.abs((data - data.mean()) / data.std())
        outliers = data[z_scores > 3]
        return outliers


class CorrelationAnalyzer:
    def calculate_correlation(self, data):
        return data.corr()

    def visualize_correlation(self, correlation_matrix):
        sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
        plt.title("Correlation Heatmap")
        plt.show()


class DataTransformer:
    def transform_data(self, data):
        #standardize the data
        standardized_data = (data - data.mean()) / data.std()
        return standardized_data


class FeatureSelector:
    def select_features(self, data):
        selected_features = data.columns
        return selected_features

print(data.head)
data_importer = DataImporter()
data_cleaner = DataCleaner()
data_describer = DataDescriber()
data_visualizer = DataVisualizer()
outlier_detector = OutlierDetector()
correlation_analyzer = CorrelationAnalyzer()
data_transformer = DataTransformer()
feature_selector = FeatureSelector()



# Step a Data Cleaning
data = data_cleaner.clean_data(data)

# Step b Data Description
summary_stats = data_describer.generate_summary_statistics(data)

# Step c Data Visualization #complete the visualizatons
data_visualizer.create_histogram(data, "Fresh")
data_visualizer.create_boxplot(data, "Milk")
data_visualizer.create_scatterplot(data, "Fresh", "Milk")
data_visualizer.create_heatmap(data)

# Step d Outlier Detection
outliers = outlier_detector.detect_outliers(data)

# Step e Correlation Analysis
correlation_matrix = correlation_analyzer.calculate_correlation(data)
correlation_analyzer.visualize_correlation(correlation_matrix)

# Step f Data Transformation
transformed_data = data_transformer.transform_data(data)

# Step g Feature Selection
selected_features = feature_selector.select_features(data)


```

# Part II - KMeans Clustering

The objective of the analysis is to group similar products together into clusters based on their attributes such as fresh, milk, grocery, frozen, detergents_paper, and delicatessen. To perform the k-means clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters, initialize the centroids, assign data points to clusters, update the centroids, and repeat until convergence.

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

class KMeansClustering:
    def __init__(self, n_clusters):
        self.n_clusters = n_clusters
        self.kmeans = KMeans(n_clusters=self.n_clusters, random_state=42)

    def fit(self, data):
        self.kmeans.fit(data)

    def predict(self, data):
        return self.kmeans.predict(data)

    def get_cluster_centers(self):
        return self.kmeans.cluster_centers_

    def get_labels(self):
        return self.kmeans.labels_



preprocessed_data = data[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']].copy()
preprocessed_data = preprocessed_data.dropna()  

# Determining the Optimal Number of Clusters
# Use the elbow method
inertia = []
k_values = range(2, 11)
for k in k_values:
    kmeans = KMeansClustering(n_clusters=k)
    kmeans.fit(preprocessed_data)
    inertia.append(kmeans.kmeans.inertia_)
plt.plot(k_values, inertia)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()


k = 5  # number of clusters
kmeans = KMeansClustering(n_clusters=k)
kmeans.fit(preprocessed_data)
centroids = kmeans.get_cluster_centers()

# Assigning Data Points to Clusters
labels = kmeans.get_labels()

# Final Results
clustered_data = preprocessed_data.copy()
clustered_data['Cluster'] = labels
print(clustered_data)
```

# Part III - Hierarchical Clustering 

Hierarchical clustering is a popular unsupervised machine learning algorithm that is used to identify patterns and group similar data points together in a hierarchy. The algorithm works by iteratively merging or splitting clusters based on a similarity measure until a dendrogram is formed.

To perform hierarchical clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters using techniques such as dendrogram.

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

class HierarchicalClustering:
    def __init__(self, data):
        self.data = data
        self.preprocessed_data = None
        self.dendrogram = None
        self.labels = None
        self.clustered_data = None
        
    def preprocess_data(self):
        self.preprocessed_data = self.data[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']].copy()
        self.preprocessed_data = self.preprocessed_data.dropna()
        
    def generate_dendrogram(self, method='ward', metric='euclidean'):
        self.dendrogram = sch.dendrogram(sch.linkage(self.preprocessed_data, method=method, metric=metric))
        
    def determine_clusters(self, n_clusters):
        self.labels = sch.fcluster(sch.linkage(self.preprocessed_data, method='ward', metric='euclidean'),
                                   n_clusters, criterion='maxclust')
        
    def assign_cluster_labels(self):
        self.clustered_data = self.preprocessed_data.copy()
        self.clustered_data['Cluster'] = self.labels
        
    def visualize_clusters(self, x, y):
        plt.scatter(self.clustered_data[x], self.clustered_data[y], c=self.clustered_data['Cluster'])
        plt.xlabel(x)
        plt.ylabel(y)
        plt.title('Hierarchical Clustering')
        plt.show()



# instantialize
hc = HierarchicalClustering(data)

# pipeline steps
hc.preprocess_data()
hc.generate_dendrogram()
hc.determine_clusters(n_clusters=2)
hc.assign_cluster_labels()
hc.visualize_clusters(x='Milk', y='Grocery')

# Generate and visualize different dendrograms 
methods = ['single', 'complete', 'average']
metrics = ['euclidean', 'cityblock', 'cosine']

for method in methods:
    for metric in metrics:
        hc.generate_dendrogram(method=method, metric=metric)
        plt.title(f"Dendrogram (Method: {method}, Metric: {metric})")
        plt.show()

```

# Part IV - PCA

In this section you are going to perform principal component analysis (PCA) to draw conclusions about the underlying structure of the wholesale customer data. Since using PCA on a dataset calculates the dimensions which best maximize variance, we will find which compound combinations of features best describe customers.

```python
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

class PCAPipeline:
    def __init__(self, n_components=2):
        self.n_components = n_components
        self.pca = PCA(n_components=self.n_components)
        self.transformed_data = None
        self.explained_variance_ratio_ = None
    
    def fit_transform(self, data):
        preprocessed_data = data[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']].copy()
        preprocessed_data = preprocessed_data.dropna()
        
        self.transformed_data = self.pca.fit_transform(preprocessed_data)
        self.explained_variance_ratio_ = self.pca.explained_variance_ratio_
        
        return self.transformed_data
    
    def plot_scree(self):
        cumulative_variance = self.explained_variance_ratio_.cumsum()
        
        plt.plot(range(1, self.n_components + 1), cumulative_variance, marker='o', linestyle='--')
        plt.xlabel('Number of Principal Components')
        plt.ylabel('Cumulative Variance Explained')
        plt.title('Scree Plot')
        plt.show()
    
    def plot_pca(self):
        plt.scatter(self.transformed_data[:, 0], self.transformed_data[:, 1])
        plt.xlabel('Principal Component 1')
        plt.ylabel('Principal Component 2')
        plt.title('PCA: Transformed Data')
        plt.show()

```

```python
# instantialize
pipeline = PCAPipeline()

# fit and transform
transformed_data = pipeline.fit_transform(data)

# have a look at transformed data
print(transformed_data)
print(pipeline.explained_variance_ratio_)

# Plot the scree plot and PCA visualization
pipeline.plot_scree()
pipeline.plot_pca()
```
